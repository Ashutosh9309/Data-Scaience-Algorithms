```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df = pd.DataFrame(data = cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
std.fit(df)
```

```python
data = std.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(data)
```

```python
X_pca = pca.transform(data)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,8))
sns.scatterplot(x = X_pca[:,0], y = X_pca[:,1], palette='plasma', hue=cancer['target'])
plt.xlabel('First Components')
plt.ylabel('Second Components')
```

```python
pca.components_
```

```python
pca_cmp = pd.DataFrame(data = pca.components_, columns=cancer['feature_names'])
```

```python
pca_cmp.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(pca_cmp, cmap='plasma')
```
